```python
# üì¶ Importy
import os
import pickle
from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# üîç Funkcja opisujƒÖca za≈Çadowany obiekt
def describe_object(obj):
    if isinstance(obj, dict):
        print(f"  ‚Üí dict with {len(obj)} keys: {list(obj.keys())[:5]}...")
    elif isinstance(obj, list):
        print(f"  ‚Üí list with {len(obj)} elements")
        if obj:
            print(f"    - type of first element: {type(obj[0])}")
    elif isinstance(obj, pd.DataFrame):
        print(f"  ‚Üí DataFrame with shape {obj.shape}")
        display(obj.head())
    elif hasattr(obj, 'shape'):
        print(f"  ‚Üí object with shape: {obj.shape}")
    else:
        print(f"  ‚Üí type: {type(obj)}")

# üìÇ Funkcja skanujƒÖca katalog i wypisujƒÖca zawarto≈õƒá plik√≥w pickle
def inspect_pickle_file(filepath):
    print(f"\nüìÇ {filepath}")
    try:
        with open(filepath, 'rb') as f:
            obj = pickle.load(f)
            describe_object(obj)
            return obj
    except Exception as e:
        print(f"  ‚ö†Ô∏è Error loading {filepath.name}: {e}")


```

```python
#!/usr/bin/env python3
import os, glob, subprocess, argparse
from pathlib import Path

# ---- LISTY ----
splits = {
    "train": [
        "case_000",
        "case_001",
        "case_002",
        "case_003",
        "case_004",
        "case_006",
        "case_007",
        "case_011",
        "case_012",
        "case_015",
        "case_016",
        "case_019",
        "case_020",
        "case_022",
        "case_024",
        "case_025",
        "case_026",
        "case_029",
        "case_030",
        "case_031",
        "case_035",
        "case_037",
        "case_038",
        "case_039",
        "case_040",
        "case_043",
        "case_046",
        "case_047",
        "case_048",
        "case_049",
        "case_050",
        "case_051",
        "case_053",
        "case_055",
        "case_056",
        "case_057",
        "case_058",
        "case_059",
        "case_060",
        "case_061",
        "case_063",
        "case_064",
        "case_065",
        "case_066",
        "case_067",
        "case_069",
        "case_070",
        "case_071",
        "case_072",
        "case_073",
        "case_075",
        "case_076",
        "case_078",
        "case_080",
        "case_081",
        "case_082",
        "case_083",
        "case_084",
        "case_085",
        "case_088",
        "case_089",
        "case_090",
        "case_091",
        "case_094",
        "case_095",
        "case_096",
        "case_097",
        "case_098",
        "case_100",
        "case_101",
        "case_102",
        "case_103",
        "case_104",
        "case_105",
        "case_108",
        "case_109",
        "case_110",
        "case_111",
        "case_113",
        "case_114",
        "case_115",
        "case_116",
        "case_120",
        "case_121",
        "case_122",
        "case_123",
        "case_125",
        "case_126",
        "case_128",
        "case_129",
        "case_133",
        "case_134",
        "case_135",
        "case_136",
        "case_138",
        "case_139",
        "case_140",
        "case_141",
        "case_142",
        "case_144",
        "case_145",
        "case_146",
        "case_147",
        "case_148",
        "case_149",
        "case_150",
        "case_151",
        "case_152",
        "case_153"
    ],
    "val": [
        "case_005",
        "case_008",
        "case_013",
        "case_014",
        "case_018",
        "case_021",
        "case_028",
        "case_042",
        "case_044",
        "case_045",
        "case_062",
        "case_086",
        "case_087",
        "case_092",
        "case_093",
        "case_107",
        "case_112",
        "case_117",
        "case_118",
        "case_127",
        "case_130",
        "case_131",
        "case_154"
    ],
    "test": [
        "case_009",
        "case_010",
        "case_017",
        "case_023",
        "case_027",
        "case_032",
        "case_033",
        "case_034",
        "case_036",
        "case_041",
        "case_052",
        "case_054",
        "case_068",
        "case_074",
        "case_077",
        "case_079",
        "case_099",
        "case_106",
        "case_119",
        "case_124",
        "case_132",
        "case_137",
        "case_143"
    ]
}

SRC = Path("/net/pr2/projects/plgrid/plgg_13/surg_vu")          
DEST_ROOT = Path("/net/pr2/projects/plgrid/plgg_13/surgvu_data")


def extract_one_tar(tar_path: Path, dest: Path, dry_run: bool):
    dest.mkdir(parents=True, exist_ok=True)
    if dry_run:
        print(f"[DRY] tar -xf {tar_path} -C {dest}")
        return
    subprocess.run(["tar", "-xf", str(tar_path), "-C", str(dest)], check=True)

def main(dry_run: bool):
    for split in splits.keys():
        (DEST_ROOT / split).mkdir(parents=True, exist_ok=True)

    for split, cases in splits.items():
        for case in cases:
            dest_case = DEST_ROOT / split / case
            done_mark = dest_case / ".extract_done"
            if done_mark.exists() and not dry_run:
                print(f"[SKIP] {split}/{case} (already done)")
                continue

            parts = sorted(glob.glob(str(SRC / f"{case}_video_part_*.tar")))
            if not parts:
                print(f"[WARN] No TARs for {case}")
                continue

            print(f"[INFO] {case}: {len(parts)} parts -> {dest_case}")
            try:
                for tarfile in parts:
                    extract_one_tar(Path(tarfile), dest_case, dry_run=dry_run)
                if not dry_run:
                    done_mark.touch()
                    print(f"[OK]   {case}")
            except subprocess.CalledProcessError as e:
                print(f"[ERR]  {case}: tar failed ({e.returncode})")
            except Exception as e:
                print(f"[ERR]  {case}: {e}")

if __name__ == "__main__":
    import argparse
    parser = argparse.ArgumentParser()
    parser.add_argument("--dry-run", action="store_true", help="Only show what would be extracted")
    args = parser.parse_args()
    main(dry_run=args.dry_run)
```
